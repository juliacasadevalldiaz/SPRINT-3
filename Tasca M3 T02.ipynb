```python
df_reduit.describe()
```

**Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).**

```python
#velocitat mitjana = distancia/temps 

df_reduit['Velocitat_mitjana'] = df_reduit['Distance'] / df_reduit['AirTime']
df_reduit
```

**Fes una taula de les aerolínies amb més endarreriments acumulats.**


```python
#agrupem el dataset df_reduit per aerolinies, despres agafem nomes la columna ArrDelay i les sumem per cada aerolinia:
endarreriment_acumulat = df_reduit.groupby('UniqueCarrier')['ArrDelay'].sum()
endarreriment = endarreriment_acumulat.sort_values(ascending=False)
pd.DataFrame(endarreriment)
print(endarreriment)
```

**Quins són els vols més llargs?**

```python
flight_num_counts = df['FlightNum'].value_counts()
print(flight_num_counts)
```

Veiem que per un mateix FlightNum hi ha diferents dades, aixi que no puc sumar sino que calculem la mitjana dels valors.

```python
temps_vol_df = df_reduit.groupby(['FlightNum', 'Origin', 'Dest'])['AirTime'].mean()
vols_mes_llargs = temps_vol_df.sort_values(ascending=False)


print(vols_mes_llargs.head())
```

**I els més endarrerits?(segons el FlighNum, es a dir, els vols que s'han fet que s'han edarrerit més)**

```python
retras = df_reduit.groupby(['FlightNum', 'Origin', 'Dest'])['ArrDelay'].mean()

vols_mes_retrasats = retras.sort_values(ascending=False)


print(vols_mes_retrasats.head())
```

**Busca les rutes més llargues**

```python
distt = df_reduit.groupby(['Origin', 'Dest'])['Distance'].mean()

rutes_dist_2 = distt.sort_values(ascending=False)


print(rutes_dist_2.head())
```

**i les que acumulen més retards.(mitjana de tots els vols d'una mateixa ruta)**

```python
distancia_vol_df = df_reduit.groupby(['Origin', 'Dest'])['ArrDelay'].mean()
vols_ordenats_per_distancia = distancia_vol_df.sort_values(ascending=False)


print(vols_ordenats_per_distancia.head())
```
